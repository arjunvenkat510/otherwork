```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
```

### Opioids (VA) graphs

```python
df_opioids = pd.read_csv('OpiodsVA.csv')
```

#### Do opioid overdoes tend to be associated with less affluent areas—that is, areas where families have lower incomes?

```python
"""A scatter plot was used since the question is asking for the relationship between two variables."""

fig, ax = plt.subplots()
ax.scatter(df_opioids['MedianHouseholdIncome'], df_opioids['FPOO-Rate'], alpha = 0.4, c = 'g')  
fig.suptitle('Opioid Overdoses vs. Median Household Income')   
ax.xaxis.set_label_text('Median Household Income') 
ax.yaxis.set_label_text('Opioid overdoses per 100,000 county residents')       
fig.set_size_inches(8, 5)
plt.show()
```

#### Some people who start with opioid addictions are reported to transition to heroin use. What is the relationship in Virginia counties between opioid overdoses and heroin overdoses?

```python
"""Box plots were used to compare the two variables side-by-side since both data series used the same
units of measurement (overdoses per 100,000 people), and this type of graph allows for an easy visual comparison."""

fig, ax = plt.subplots()
overdoses = []  # creates an empty list
overdoses.append(df_opioids['FFHO-Rate'])  # appends the opioid overdoses to the list
overdoses.append(df_opioids['FPOO-Rate'])  # appends the heroin overdoses to the list
ax.boxplot(overdoses, labels = ['Opioids','Heroin'])
fig.suptitle('Opioid overdoses vs. Heroin overdoses in Virginia counties')
ax.xaxis.set_label_text('Drug Type')
ax.yaxis.set_label_text('Overdoses per 100,000 county residents')
fig.set_size_inches(8, 8)
plt.show()
```

### Presidents

```python
df_presidents = pd.read_csv('presidents.csv')
```

#### Which states are associated with the greatest number of United States presidents in terms of the presidents’ birthplaces?

```python
"""A histogram was used for this data set since the question asks for the frequency of the data; only the 7 
most popular birthplace states were included since the remaining 14 states had one president born in each one.""" 

states = df_presidents['State']
states = dict(states.value_counts())
statenames = [i for i in states.keys()]   # Converts the state names into a list
frequency = [i for i in states.values()]  # Converts the frequency into a list
fig, ax = plt.subplots()
ax.bar(x = statenames[:7], height = frequency[:7], color = 'r')  #[:7] to only include the first seven states
fig.suptitle("Location of U.S. Presidents' birthplace")
fig.set_size_inches(12, 5)
ax.xaxis.set_label_text('State')
ax.yaxis.set_label_text('Number of Presidents')
plt.show()
```

### United States vehicle sales

```python
df_cars = pd.read_csv('TOTALNSA.csv')
```

#### How have vehicle sales in the United States varied over time?

```python
"""A line graph was used since this is continuous data (time series) with the line implying a connection 
between the data points."""

fig,ax = plt.subplots()
ax.plot(df_cars['DATE'], df_cars['TOTALNSA'])
fig.suptitle('Total U.S. monthly vehicle sales from 1976 - 2018')
ax.xaxis.set_label_text('Year-Month-Day')
ax.yaxis.set_label_text('Total U.S. Sales')
fig.set_size_inches(15, 5)
plt.show()
```
